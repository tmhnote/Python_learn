### 68 字符串的处理-字符串的拼接操作

拼接方式

使用str.join()方法进行拼接字符串

直接拼接

使用格式化字符串进行拼接

```python
# 示例 6-7 字符串的拼接操作
s1 ='hello'
s2 ='world'
```

```python
# (1) 使用 + 直接拼接
print(s1+s2)
```

```python
# (2) 使用字符串的join()方法
print(''.join([s1,s2])) # 使用空字符串 进行拼接

# 使用 * 进行拼接
print('*'.join(['hello','world', 'python', 'java', 'php']))
print('你好'.join(['hello','world', 'python', 'java', 'php']))
```

```python
# (3) 直接拼接，写在一起
print('hello''world')
```

```python
# (4)使用格式化字符串进行拼接
print('%s%s'%(s1,s2))
print(f'{s1}{s2}')
print('{0} {1}'. format(s1,s2))
```

### 68 字符串的处理-字符串的去重操作

```python
# 示例6-8
s = 'helloworldhelloworldhelloworldadfff'
```

```python
# (1)字符串拼接及not in
new_s = ''
for item in s:
    if item not in new_s:
        new_s += item
print(new_s)
```

```python
# (2)使用索引 + not in 
new_s2 = ''
for i in range(len(s)):
    if s[i] not in new_s2:
        new_s2 += s[i]
print(new_s2)
```

```python
# (3) 通过集合去重 + 列表排序
new_s3 = set(s)
lst = list(new_s3)
print(lst)
lst.sort(key=s.index)
print(''.join(lst))
```
